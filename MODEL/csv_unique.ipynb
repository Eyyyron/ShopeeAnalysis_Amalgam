```python
import pandas as pd

input_file = "sarima_residuals_top_second.csv"
output_file = "unique_categories.csv"

# Read CSV (first two columns only)
df = pd.read_csv(input_file, usecols=[0, 1])

# Drop duplicate rows based on these two columns
df_unique = df.drop_duplicates()

# Save the cleaned result
df_unique.to_csv(output_file, index=False)

print("Unique rows saved", output_file)
```
